## 머신러닝 알고리즘
- 사이킷런은 유명한 머신러닝용 라이브러리이다.
- 머신러닝의 알고리즘 종류는 크게 3가지로 나눌 수 있다.
  1. 지도학습(Supervised Learning)
  2. 비지도학습(Unsupervised Learning)
  3. 강화학습(Reinforcement Learning)
- 위 3가지 카테고리는 상황이나 구현하고자 하는 Application에 따라 합쳐서 사용될 수도 있다.

### 알고리즘 선택 요령
- 보통 라벨(정답)의 존재 유무에 따라 머신러닝을 지도학습과 비지도학습으로 나눈다.
- 데이터가 복잡하다면 차원을 축소하기도 하고 주요한 성분이 있다면 주성분 분석을 적용하기도 한다.
- 데이터의 종류에 따라 같은 회귀라도 선형회귀는 수치형 데이터를 예측하고 로지스틱 회귀는 분류(이진 분류)에 사용된다.
- 이와 같이 정답 유무, 데이터의 종류, 특성, 문제에 따라 머신 러닝 알고리즘을 선택해서 사용해야 한다.

### 지도학습(Supervised learning)
- 지도학습 알고리즘은 한 세트의 사례들을 기반으로 예측을 수행한다.
- 분류(Classification) : 데이터가 범주형(categorical) 변수를 예측하기 위해 사용될 때 지도학습을 '분류'라고 부른다. 레이블이 두 개인 경우를 '이진 분류(binary classification)'라고 부르며 범주가 두 개 이상인 경우 다중 클래스 분류(multi-class classification)라고 부른다.
- 회귀(Regression) : 연속되는 값을 예측해야 하는 문제를 회귀라고 한다.
- 예측(Forecasting) : 과거 및 현재 데이터를 기반으로 미래를 예측한다. 예측은 동향(trends)를 분석하기 위해 가장 많이 사용된다.

### 비지도학습(Unsupervised learning)
- 비지도학습을 수행할 때 기계는 미분류 데이터를 제공 받는다. 
- 그리고 제공 받은 클러스터링 구조, 저차원 다양체, 희소 트리 및 그래프 등과 같은 데이터를 통해서 고유 패턴을 발견하도록 설정된다.
- 클러스터링(Clustering) : 특정 기준에 따라 유사한 데이터 사례들을 하나의 세트로 그룹화한다. 이 과정은 전체 데이터 세트를 여러 그룹으로 분류하기 위해 사용된다. 사용자는 고유한 패턴을 찾기 위해 개별 그룹 차원에서 분석을 수행할 수 있다.
- 차원 축소(Dimension Reduction) : 고려 중인 변수의 개수를 줄이는 작업을 의미한다. 많은 어플리케이션에서 원시 데이터는 아주 높은 차원의 특징을 가지고 있는데 이때 일부 특징은 중복되거나 작업과 아무 관련이 없다. 따라서 차원수를 줄이면 작업을 진행하는데 수월해진다.

### 강화학습(Reinforcement learning)
- 강화학습은 환경으로부터의 피드백을 기반으로 행위자(agent)의 행동을 분석하고 최적화한다.
- 최고의 보상을 산출하는 액션을 발견하기 위해 서로 다른 시나리오들을 시도한다.

### 데이터 표현법
- 사이킷런에서는 데이터 표현 방식을 보통 2가지로 나타낸다.
- 특성 행렬(Feature Matrix)와 타겟 벡터(Target Vector)이다.
image.png

#### 특성 행렬(Feature Matrix)
- 입력 데이터를 의미한다.
- 특성(feature)는 데이터에서 수치 값, 이산 값, 불리언 값으로 표현되는 개별 관측치를 의미한다. 특성 행렬에서는 열에 해당하는 값이다.
- 표본은 각각의 입력 데이터로 특성 행렬에서는 행에 해당하는 값이다.
- n_sample은 행의 개수(표본의 개수)를 의미한다.
- n_features는 열의 개수(특성의 개수)를 의미한다.

#### 타겟 벡터(Target Vector)
- 입력 데이터의 라벨(정답)을 의미한다.
- 목표(Target)는 라벨, 타겟값, 목표값이라고 부르며 특성 행렬(Feature Matrix)로부터 예측하고자 하는 것을 의미한다.
- n_sample은 벡터의 길이(라벨의 개수)를 의미한다.
- 타겟 벡터에서는 n_features는 없다.
- 타겟 벡터는 보통 1차원 벡터로 나타내며 ndarray, Series를 사용하여 나타낼 수 있다.

## 사이킷런 주요 모듈(1) 회귀 모델 실습

```python
import numpy as np
import matplotlib.pyplot as plt
r = np.random.RandomState(10)
x = 10 * r.rand(100)
y = 2 * x - 3 * r.rand(100)
plt.scatter(x,y)
```

```python
# 입력 데이터 x 확인

x.shape
```

```python
# 정답 데이터 y 확인

y.shape
```

```python
# 모델 객체 생성 - LinearRegression 사용

from sklearn.linear_model import LinearRegression
model = LinearRegression()
model
```

```python
# 모델 훈련

# ! 에러 발생
# model.fit(x, y)

#이렇게 하면 오류가 발생하는데 그 이유는 입력 데이터 x를 행렬로 바꾸지 않아서 오류가 발생한다.
```

```python
# 입력 데이터 x를 행렬로 변환

X = x.reshape(100, 1)
```

```python
model.fit(X, y)
```

```python
# 입력데이터와 라벨을 이용해서 훈련이 끝난 모델에 새로운 데이터를 넣고 예측하기

x_new = np.linspace(-1, 11, 100)
X_new = x_new.reshape(100,1)
y_new = model.predict(X_new)
```

```python
# 모델 평가

from sklearn.metrics import mean_squared_error

error = np.sqrt(mean_squared_error(y,y_new))

print(error)
```

```python
plt.scatter(x, y, label='input data')
plt.plot(X_new, y_new, color='red', label='regression line')
```
